```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
latency = np.loadtxt("test.txt")

latency = latency[latency < 5000]
mean_latency = np.mean(latency)

plt.plot(latency)
plt.xlabel('Frames')
plt.ylabel('latency (ms)')
plt.title('Latency (Avg. Latency = ' + str(round(mean_latency)) + ' ms)')
plt.savefig('Latency.pdf')
plt.show()
```

```python
# Python
with open('test.txt', 'r') as f:
    lines = f.readlines()

# NumPy
numbers = np.empty((len(lines), 2))

# 
for i, line in enumerate(lines):
    # 
    x, y = line.strip('()\n').split(',')
    
    # NumPy
    numbers[i] = [float(x), float(y)]
```

```python
numbers = numbers[numbers[:,0] <= numbers[:,1]]
numbers = numbers[numbers[:,0] > 5000000]
print(numbers)
```

```python
delta = numbers[:,1] - numbers[:,0]
print(delta)
```
